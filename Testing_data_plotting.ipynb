# Working with Data Files

## Working with CSV files

More details [here](https://www.shanelynn.ie/python-pandas-read-csv-load-data-from-csv-files/).

To read data files, you need to be within the same working directory as that data file. Module `os` is useful to:
- find the current working directory (required data file can be moved there)
- change your current working directory to the directory where the files reside using ``os.chdir()``

Pandas is required for reading cvs file type.

```python
import os
import pandas as pd
```

```python
# Find out your current working directory
print(os.getcwd())

# Display all of the files found in your current working directory
print(os.listdir(os.getcwd()))
```

When specifying file names to the read_csv function, you can supply both absolute or relative file paths:
- **relative path** - starting from your current working directory
- **absolute path** - complete path from the base of your file system to the file that you want to load, e.g. c:/Documents/Shane/data/test_file.csv

> ! recommended and preferred to **use relative paths** where possible in applications, because absolute paths are unlikely to work on different computers due to different directory structures !

You can change the delimiter (depending what used in the file) using `sep` keyword.

```python
#load file using relative path
pd.read_csv('eddypro_FdP_full_output_2022-01-28T073241_adv.csv', sep=';')
#load file using absolute path
#pd.read_csv('C:\Users\Stela Kusar\Code\Python\BTR\eddypro_FdP_full_output_2022-01-28T073241_adv.csv', sep=';')
```

There are any other keywords, which allow a more customized/modified data to be loaded. You can check options with `dir(pd.read_csv)`.

Note that the header parameter was set to `True` by default. That means the method automatically detects and assigns the first row of the CSV file as the pandas dataframe header. If you didn’t have a header in your csv data, you would want to set `header=None`.

```python
#Customize loaded data

data = pd.read_csv('eddypro_FdP_full_output_2022-01-28T073241_adv.csv',   
                   sep=';',           #semicolon value file
                   usecols=[1,2,10,11,13,14],   #only load the specified columns
                   parse_dates=[0],     # Intepret the 1st column as a date
                   skiprows=[0,2,3],         #skip the 3rd row of the file
                   na_values=['NaN'],       #take any 'NaN' as missing values
                   dtype={"qc_H":int},             #parse the salary column as an integer
                   #MORE KEYWORDS
                   #quotechar="'",        #single quote allowed as quote character
                   
)
print(data.shape)

data1 = data.dropna() #filter out all rows containing one or more missing values
print(data1.shape)

data.head(10) #only show first 10 rows
```

Since we only want to wor and plot with certain subset of our data (of best quality), we need to filter what we want to keep. We can do that using boolean variables (``x=?``) or expressions (`x.eg(?)`).

Learn more how to [acces pandas dataframe](https://pythonhow.com/python-tutorial/pandas/Accessing-pandas-dataframe-columns-rows-and-cells/) and how to [filter out specific values](https://cmdlinetips.com/2018/02/how-to-subset-pandas-dataframe-based-on-values-of-a-column/).

```python
print("Before filtering:",data.shape)

#Filter rows with qc_H=0 using a boolean variable
is_0 =  data['qc_H']==0 #boolean variable 
data_H = data[is_0]

print("After filtering:",data_H.shape)
```

```python
#Filter rows using the boolean expression
data_H = data[data.qc_H.eq(0)]
#OR data_H = data[data["qc_H"].eq(0)]

print(data_H.shape)
```

```python
#Exclude data of a specific value
data_H = data[data["qc_H"] != 1]
print(data_H.shape)

#Select rows whose column value is in the list 
qc = [1, 2]
data_H = data[data.qc_H.isin(qc)]
print(data_H.shape)

#Select rows whose column value is NOT in the list
data_H = data[~data.qc_H.isin(qc)]
print(data_H.shape)
```

```python
#Filter out rows in dataframe with column year values NA/NAN
data_no_NA = data[data.LE.notnull()]
print(data_no_NA.shape)

#Combine conditions for the selection
data_LE = data[data.LE.notnull() & data.qc_LE.eq(0)] #using logical expressions
print(data_LE.shape)

data_HLE = data.query('qc_LE==0 & qc_H == 0') #using query() funtion
print(data_HLE.shape)
```

We can make sure our new data frame contains only rows corresponding to specified values:

```python
data_H = data[data.qc_H.eq(0)]
data.qc_H.unique(),data_H.qc_H.unique()
```
